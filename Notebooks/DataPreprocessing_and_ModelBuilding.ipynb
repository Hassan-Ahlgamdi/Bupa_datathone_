## EDA and some Charts

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_parquet('final_claims_dataset.parquet')
```

```python
# 1. Prepare the data: Group by region, calculate the mean, and sort
avg_claims_by_region = df.groupby('PROVIDER_REGION')['CLAIMS_AMOUNT'].mean().sort_values(ascending=False).reset_index()

# 2. Create the plot
plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    x='PROVIDER_REGION',
    y='CLAIMS_AMOUNT',
    data=avg_claims_by_region,
    palette=['lightgrey'] * len(avg_claims_by_region) # Start with all bars grey
)

# 3. Highlight the 'Overseas' bar
# Find the index of the 'Overseas' region in our sorted data
try:
    overseas_index = avg_claims_by_region[avg_claims_by_region['PROVIDER_REGION'] == 'Overseas'].index[0]
    # Change the color of that specific bar
    barplot.patches[overseas_index].set_facecolor('salmon')
except IndexError:
    print("Note: 'Overseas' category not found in the data.")


# 4. Add titles and labels for clarity
plt.title('Average Claim Amount by Provider Region', fontsize=16)
plt.xlabel('Provider Region', fontsize=12)
plt.ylabel('Average Claim Amount', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to make sure everything fits

# 5. Save the figure to a file
plt.savefig('average_claim_by_region.png')

print("Chart has been saved as 'average_claim_by_region.png'")
```

```python
df = pd.read_parquet('final_df.parquet')

model_columns = ['STATUS', 'GENDER',
       'AGE', 'CLAIM_TYPE',
       'COUNT_DISTINCT_of_VOU_NO',
    'PREAUTH_AMOUNT',
    'NUMBER_OF_PA_EPISODES',
        'HAS_PREAUTH',
    'PLAN_NETWORK_MAPPED',
        'DIAG_CODE_COUNT_LOG',
    'IS_COMMON_DIAG',
        'CLAIMS_AMOUNT_LOG',
        'PROVIDER_REGION']
model_df = df[model_columns]
model_df
```

```python
model_df_encoded = pd.get_dummies(model_df, columns=['STATUS', 'GENDER', 'CLAIM_TYPE', 'PROVIDER_REGION'])
model_df_encoded
```

```python
model_df_encoded.columns
```

```python
encoded_columns = [
    'STATUS_AC', 'STATUS_RJ', 'STATUS_SU', 'GENDER_F', 'GENDER_M',
       'CLAIM_TYPE_I', 'CLAIM_TYPE_O', 'PROVIDER_REGION_Central',
       'PROVIDER_REGION_Eastern', 'PROVIDER_REGION_Northern',
       'PROVIDER_REGION_Overseas', 'PROVIDER_REGION_Southern',
       'PROVIDER_REGION_Western'
]
model_df_encoded[encoded_columns] = (model_df_encoded[encoded_columns] == True).astype(int)
model_df_encoded
```

```python
model_df_encoded.to_parquet(path = 'model_df_encoded.parquet')
```

```python
## Linear Regression
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import mean_squared_error as mse, mean_absolute_error as mae, r2_score, accuracy_score
model_df_encoded = pd.read_parquet('model_df_encoded.parquet')
X = model_df_encoded.drop('CLAIMS_AMOUNT_LOG', axis=1)
y = model_df_encoded['CLAIMS_AMOUNT_LOG']

# 1. Initialize the model
model = LinearRegression()

# 2. Set up the K-Fold cross-validator
# n_splits=5 means we'll do 5 folds (a common choice)
# shuffle=True is important to randomize the data before splitting
# random_state ensures the shuffle is the same every time you run the code
kfolds = KFold(n_splits=5, shuffle=True, random_state=42)

# 3. Perform cross-validation and calculate scores
# We'll calculate scores: Mean Squared Error and R-squared and Mean Absolute Error

mses = []
maes = []
r2s = []

# 4. Training loop
for fold, (train_index, test_index) in enumerate(kfolds.split(X,y)):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # train the model once
    model.fit(X_train, y_train)

    # Make predictions
    y_pred = model.predict(X_test)

    # Calculate metrics
    # --- Calculate and store metrics ---
    mses.append(mse(y_test, y_pred))
    maes.append(mae(y_test, y_pred))
    r2s.append(r2_score(y_test, y_pred))

    print(f"Fold {fold+1}:  MSE = {mses[fold]:.4f}, MAE = {maes[fold]},  R2 = {r2s[fold]:.4f}" )


print(f" AVERAGE RMSE = {np.sqrt(np.mean(mses)):.4f}, AVERAGE MAE = {np.mean(maes)}, AVERAGE R2 = {np.mean(r2s):.4f}" )
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, mean_absolute_error, mean_squared_error


X = model_df_encoded.drop('CLAIMS_AMOUNT_LOG', axis=1)
y = model_df_encoded['CLAIMS_AMOUNT_LOG']

# 1. Initialize the model
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
poly_model = LinearRegression()

# 2. Set up the K-Fold cross-validator
# n_splits=5 means we'll do 5 folds (a common choice)
# shuffle=True is important to randomize the data before splitting
# random_state ensures the shuffle is the same every time you run the code
kfolds = KFold(n_splits=5, shuffle=True, random_state=42)

# 3. Perform cross-validation and calculate scores
mae_scorer = make_scorer(mean_absolute_error)
rmse_scorer = make_scorer(lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true, y_pred)))


r2_scores = cross_val_score(poly_model, X_poly, y, cv=kfolds, scoring='r2')
mae_scores = cross_val_score(poly_model, X_poly, y, cv=kfolds, scoring=mae_scorer)
rmse_scores = cross_val_score(poly_model, X_poly, y, cv=kfolds, scoring=rmse_scorer)


print("🔺 Cross-Validation Results (Polynomial + Interaction Linear Regression)")
print("-------------------------------------------------------------------")
print(f"R² per fold:   {r2_scores}")
print(f"Mean R²:       {r2_scores.mean():.4f}")

print(f"\nMAE per fold:  {mae_scores}")
print(f"Mean MAE:      {mae_scores.mean():.4f}")

print(f"\nRMSE per fold: {rmse_scores}")
print(f"Mean RMSE:     {rmse_scores.mean():.4f}")
```

```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

model_df_encoded = pd.read_parquet('model_df_encoded.parquet')
X =  model_df_encoded.drop('CLAIMS_AMOUNT_LOG', axis=1)
y = model_df_encoded['CLAIMS_AMOUNT_LOG']



xgb_model = xgb.XGBRegressor(objective='reg:squarederror', device='cuda',  n_estimators=50, random_state=42, n_jobs=-1)
k_folds = KFold(n_splits=5, shuffle=True, random_state=42)

# Lists to store scores
xgb_rmse_scores = []
xgb_r2_scores = []
xgb_mae_scores = []

print("\n--- Training XGBoost Regressor  with 50 n_estimatores ---")
for fold, (train_index, test_index) in enumerate(k_folds.split(X, y)):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    xgb_model.fit(X_train, y_train)
    y_pred = xgb_model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    xgb_rmse_scores.append(rmse)
    xgb_r2_scores.append(r2)
    xgb_mae_scores.append(mae)

    print(f"Fold {fold+1}: MSE = {mse:.4f}, R2 = {r2:.4f}, MAE = {mae:.4f}")
print(f" AVERAGE RMSE = {np.mean(xgb_rmse_scores):.4f}, AVERAGE MAE = {np.mean(xgb_mae_scores)},   AVERAGER2 = {np.mean(xgb_r2_scores):.4f}" )
```
